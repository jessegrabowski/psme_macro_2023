# Closed Economy AS-AD Model

Just like in the IS-TR-IFM notebook, we want to write a bunch of code to compute the AS-AD system for us. Here's the system we're going to work with:

$$\begin{align}
Y &= b_0 + b_1 G + b_2 r \\
i &= \bar i + \alpha(\pi - \bar{\pi}) + \beta Y \\
i &= r + \mathbb E [\pi] \\
\pi &= \mathbb E[\pi] + f(U - \bar{U}) + \varepsilon_\pi \\
U - \bar U &= gY
\end{align}
$$

## Sympy for symbolic derivations

Sympy is a python package that lets you solve symbolic expressions inside a Python program. It also lets you "compile" algebraic expression into numpy functions. We will use sympy in this notebook to avoid having to solve all of the functions ourselves, by hand.

To begin, we import the sympy library.

```python
import sympy as sp
```

Next we need to create symbolic variables. This is done using the `sp.symbols` function. We can pass a list of variable names, separate by spaces, and sympy will automatically break it into individual symbols.

```python
b0, b1, b2 = sp.symbols('b0 b_1 b_2')
Y, G, r = sp.symbols('Y G r')
```

Let's look at what we can do. Running a cell with a variable gives us back the variable as latex. Here's $r$:

```python
r
```

We can also get back expressions. Here I'm setting the first equation to zero: $Y - (b_0 + b_1 G + b_2 r) = 0$. Sympy always expects equations to be written this way.

```python
eq_1 = Y - (b0 + b1 * G + b2 * r)
eq_1
```

Let's make the rest of these equations. First create symbols:

```python
# Parameters
# You can write latex for the symbol name, but you have to put a little "r" at the start of the string
f, g, alpha, beta, E_pi, i_bar, pi_bar, U_bar, eps_pi = sp.symbols(r'f g alpha beta \mathbb{E}[\pi] \bar{i} \bar{\pi} \bar{U} \varepsilon_\pi')

# Variables
i, U, pi = sp.symbols('i U pi')
```

We can gather all of the equations into a single list. That's good for organization:

```python
equations = [
    Y - (b0 + b1 * G + b2 * r),
    i - (i_bar + alpha * (pi - pi_bar) + beta * Y),
    i - (r + E_pi),
    pi - (E_pi + f * (U - U_bar) + eps_pi),
    U - U_bar - g * Y
]

for eq in equations:
    display(eq)
```

### Directly Solve for the equlibrium

Now the main event. We can use `sp.solve` to solve a system of equations for a list of variables. We want to solve this system for $Y$ and $\pi$, but actually we have a system of 5 equations and 5 unknowns: $Y, \pi, i, r, U$. 

When doing computer algebra, it's best to always attack square systems. If you have more equations than unknowns, the system isn't "exactly determined". The solution algorithm might not be smart enough to reduce away equations. Here are the 5 solutions it comes up with:

```python
solutions = sp.solve(equations, [Y, r, i, U, pi])

solutions[Y]
```

```python
solutions[pi]
```

### Back up, solve for AS and AD

These are the equlibrium expressions we did so much work to get! 

But what if we want the reduced form equations? We can do subsitutions with the `.subs` method: 

#### AD 

```python
is_curve = equations[0]
is_curve
```

```python
# Substitute r = r - E[pi]
is_curve = is_curve.subs({r:i - E_pi})
is_curve
```

```python
# Substitute the TR curve for i
# First, solve the TR curve for i
tr_curve = equations[1]
i_expr = sp.solve(tr_curve, i)[0]
i_expr
```

```python
is_curve = is_curve.subs({i:i_expr})
is_curve
```

```python
# Lastly, solve for Y
f_Y_AD = sp.solve(is_curve, Y)[0]
f_Y_AD
```

```python
# We can also flip this around, just remember to subtract Y (because f_Y is only the RHS of the equation, the LHS is Y = f_Y!)
f_pi_AD = sp.solve(f_Y_AD - Y, pi)[0]
f_pi_AD
```

#### AS

AS is comparatively simple, we just use `subs` and `solve`.

```python
phillips_curve, okins_law = equations[3], equations[4]

# We're allowed to solve for expressions too. Solve Okin's law for U - U_bar
U_m_U_bar = sp.solve(okins_law, U - U_bar)[0]
U_m_U_bar
```

```python
# And subsitute it into the phillips curve
phillips_curve = phillips_curve.subs({U - U_bar: U_m_U_bar})
phillips_curve
```

```python
# And solve for pi
f_pi_AS = sp.solve(phillips_curve, pi)[0]
f_pi_AS
```

### A quick sanity check

Subbing AS in to AD should give the equlibrium of the system. We can check that this is the same as what we got automatically by solving the square system for all variables

```python
# set AS = AD and solve for Y
Y_solution = sp.solve(f_pi_AS - f_pi_AD, Y)[0]
Y_solution
```

```python
# it's the same!
solutions[Y]
```

```python
pi_solution = sp.solve(Y_solution - f_Y_AD, pi)[0]
```

```python
solutions[pi]
```

```python
# so is this!
pi_solution
```

## Compiling to Numpy

Now that we've solved AS, AD, and the equlibrium, we want to make plots and run simulations. 

To do this, we use a sympy function called `lambdify`. This function takes a list of input symbols, called `args`, and a sympy equation (called `expr`) and returns a python function. The function will take numbers for the list of symbols we provide, do the computation of `expr`, and return the result.

Let's look at a simpler example. The desired demand function needs 6 inputs: Y, G, r, b0, b1, b2. We can turn it into a function:

```python
eq_1
```

```python
f_eq_1 = sp.lambdify([Y, G, r, b0, b1, b2], eq_1)
```

```python
# Plug in some numbers
f_eq_1(1, 1, 1, 1, 0.2, -0.8)
```

```python
# We can also give arrays!
Y_grid = np.linspace(0, 1, 100)
f_eq_1(Y_grid, 1, 1, 1, 0.2, -0.8)
```

```python
f_pi_AS
```

```python
f_AS_curve = sp.lambdify([Y, f, g, E_pi, eps_pi], f_pi_AS)
```

```python
f_AS_curve(Y_grid, 1, 1, 1.1, 0)
```

### Compile AD Curve

```python
f_pi_AD
```

```python
f_AD_curve = sp.lambdify([Y, G, b0, b1, b2, alpha, beta, pi_bar, i_bar, E_pi], f_pi_AD)
```

```python
f_AD_curve(Y_grid, 1, 0, 0.8, -0.3, 1.5, 0.5, 1, 1, 1.1)
```

# Open Economy

Next we turn to analysis of the open economy model. To do so, we will need to decide on the exchange rate regime. First, though, let's define the open economy model:

$$\begin{align}
Y &= b_0 + b_1 G + b_2 r + b_3 Y^\star + b_4 \sigma \\
i &= \bar i + \alpha(\pi - \bar{\pi}) + \beta Y \\
i &= r + \mathbb E [\pi] \\
\sigma &= \bar{\sigma}(1 + \pi) \\
\pi &= \mathbb E[\pi] + f(U - \bar{U}) + \varepsilon_\pi \\
U - \bar U &= gY
\end{align}
$$

Several changes are made relative to the closed economy model:

1. We add international demand for exports and the nominal exchange rate ($\sigma$) to the desired demand equation
2. We add an equation that links the exchange rate to the *equilibrium* exchange rate, $\bar{\sigma}$, to inflation.

We will make some more changes depending on the exchange rate regime, but this is out basic setup.

## Fixed Exchange Rates

When exchange rates are fixd, the monetary policy has no power, so we can ignore the Taylor Rule entirely. We add an interest parity equation in it's place:

$$i = i^\star$$

Then solve for the AD curve

```python
b0, b1, b2, b3, b4 = sp.symbols('b0 b_1 b_2 b_3 b_4')
Y, Y_star, G, r = sp.symbols('Y Y^\star G r')
f, g, alpha, beta, E_pi, i_bar, pi_bar, U_bar, eps_pi = sp.symbols(r'f g alpha beta \mathbb{E}[\pi] \bar{i} \bar{\pi} \bar{U} \varepsilon_\pi')
sigma, sigma_bar = sp.symbols(r'\sigma \bar{\sigma}')
i, i_star, U, pi = sp.symbols('i i^\star U pi')
```

```python
equations = [
    Y - (b0 + b1 * G + b2 * r + b3 * Y_star + b4 * sigma),
    i - i_star, # interest parity
    i - (r + E_pi),
    sigma - sigma_bar * (1 + pi),
    pi - (E_pi + f * (U - U_bar) + eps_pi),
    U - U_bar - g * Y
]
```

```python
solutions = sp.solve(equations, [Y, r, i, U, pi, sigma])
```

```python
solutions[Y].collect(E_pi)
```

```python
solutions[pi].collect(E_pi).collect(f*g)
```

### AD Curve

To get the AD curve, make some substitutions

```python
AD_curve  = equations[0]
f_Y_AD = sp.solve(AD_curve.subs({r: i - E_pi, sigma: sigma_bar * (1 + pi)}).subs({i:i_star}),
                  Y)[0].collect(b2).collect(b4 * sigma_bar)
f_pi_AD = sp.solve(Y - f_Y_AD, pi)[0]
f_pi_AD
```

### AS Curve

The AS curve doesn't change when we open the economy!

```python
f_pi_AS = f * g * Y + E_pi + eps_pi
f_pi_AS
```

### Compile and plot

```python
# Uncomment and run this line to check what symbols are in the expression.
# It's helpful when you're writing the lambdify expressions!

# f_pi_AS.args
```

```python
f_AS_curve = sp.lambdify([Y, f, g, E_pi, eps_pi], f_pi_AS)
f_AD_curve = sp.lambdify([Y, G, sigma_bar, E_pi, Y_star, i_star, b0, b1, b2, b3, b4], f_pi_AD)
f_pi_eq = sp.lambdify([G, sigma_bar, E_pi, Y_star, i_star, eps_pi, b0, b1, b2, b3, b4, f, g], solutions[pi])
f_Y_eq = sp.lambdify([G, sigma_bar, E_pi, Y_star, i_star, eps_pi, b0, b1, b2, b3, b4, f, g], solutions[Y])
```

```python
b0 = 1 # demand constant
b1 = 1.1 # government spending multiplier
b2 = -0.3 # sensitivity of investment to real interest rate
b3 = 0.5 # sensitivty of GDP gap to foreign growth (foreign wealth effect)
b4 = -0.2 # sensitivity of GDP gap to exchange rate (trade balance sensitivity to price changes)

G = 0 # government spending
alpha = 1.5 # central bank sensitivity to excess inflation
beta = 0.8 # central bank sensitivity to excess growth
pi_bar = 2 # central bank inflation target
i_bar = 1 # central bank "base" interest
E_pi = 1 # Expected future inflation

Y_star = 1 # Foreign GDP
i_star = 1 # Foreign interest rate

sigma_bar = 1 # Fixed exchange rate

f = 1 # Phillip's curve parameter
g = 2 # Okun's law parameter
pi_shock = 0 # Supply shock to inflation 
```

### Effect of a supply shock

```python
# Compute the curves and the equlibria with the "default" parameters
# (I made them up randomly)
AD_curve = f_AD_curve(Y_grid, G, sigma_bar, E_pi, Y_star, i_star, b0, b1, b2, b3, b4)
AS_curve = f_AS_curve(Y_grid, f, g, E_pi, pi_shock)
Y_eq_val = f_Y_eq(G, sigma_bar, E_pi, Y_star, i_star, pi_shock, b0, b1, b2, b3, b4, f, g)
pi_eq_val = f_pi_eq(G, sigma_bar, E_pi, Y_star, i_star, pi_shock, b0, b1, b2, b3, b4, f, g)

# # Set a new value for the shock, and compute everything again
new_pi_shock = 1.1
AD_curve_shock = f_AD_curve(Y_grid, G, sigma_bar, E_pi, Y_star, i_star, b0, b1, b2, b3, b4)
AS_curve_shock = f_AS_curve(Y_grid, f, g, E_pi, new_pi_shock)
Y_eq_val_shock = f_Y_eq(G, sigma_bar, E_pi, Y_star, i_star, new_pi_shock, b0, b1, b2, b3, b4, f, g)
pi_eq_val_shock = f_pi_eq(G, sigma_bar, E_pi, Y_star, i_star, new_pi_shock, b0, b1, b2, b3, b4, f, g)
```

```python
# Make the plots
fig, ax = plt.subplots()
ax.plot(Y_grid, AD_curve, c='tab:blue', label = 'AD Initial')
ax.plot(Y_grid, AS_curve, c='tab:orange', label = 'AS Initial')
ax.scatter(Y_eq_val, pi_eq_val, zorder=10)

ax.plot(Y_grid, AD_curve_shock, c='tab:blue', ls='--', label = 'AD Shock')
ax.plot(Y_grid, AS_curve_shock, c='tab:orange', ls='--', label = 'AS Shock')
ax.scatter(Y_eq_val_shock, pi_eq_val_shock, zorder=10)

ax.set(xlabel='$Y$', ylabel='$\pi$', title='AS-AD System: Supply Shock')
ax.legend()
plt.show()
```

## Flexible Exchange Rates

This is almost the same, but we bring the Taylor Rule back in, and drop the expression that pinned down the exchange rate. 

```python
b0, b1, b2, b3, b4 = sp.symbols('b0 b_1 b_2 b_3 b_4')
Y, Y_star, G, r = sp.symbols('Y Y^\star G r')
f, g, alpha, beta, E_pi, i_bar, pi_bar, U_bar, eps_pi = sp.symbols(r'f g alpha beta \mathbb{E}[\pi] \bar{i} \bar{\pi} \bar{U} \varepsilon_\pi')
sigma, sigma_bar = sp.symbols(r'\sigma \bar{\sigma}')
i, i_star, U, pi = sp.symbols('i i^\star U pi')
```

```python
equations = [
    Y - (b0 + b1 * G + b2 * r + b3 * Y_star + b4 * sigma),
    i - (i_bar + alpha * (pi - pi_bar) + beta * Y),
    i - i_star, # interest parity
    i - (r + E_pi),
    pi - (E_pi + f * (U - U_bar) + eps_pi),
    U - U_bar - g * Y
]
```

```python
solutions = sp.solve(equations, [Y, r, i, U, pi, sigma])
```

Important: Y and pi do **not** depend on fiscal policy terms! 

```python
solutions[Y].collect(alpha)
```

```python
solutions[pi]
```

Only sigma does. Just like the IS-TR-IFM analysis, fluctuations in the exchange rate cause foreign demand to perfectly counterbalance any shocks in domestic demand.

```python
solutions[sigma].collect(b4).collect(E_pi).collect(alpha * f * g).collect(beta).collect(-G * b1 - Y_star * b3 - b0 - b2 * i_star)
```

For the actual curves, it's deceptively simple. The TR curve itself is the AD curve -- just solve for Y

```python
equations[1]
```

```python
sp.solve(equations[1].subs({i:i_star}), pi)[0].simplify()
```

While the AS curve is as usual

```python
f_pi_AS
```

For plotting, we also need to know how sigma varies with Y. This is just a re-arrangement of the IS curve.

```python
sigma_curve = sp.solve(equations[0].subs({r: i_star - E_pi}), sigma)[0]
sigma_curve.atoms()
```

### Compile curves into code

```python
solutions[sigma].atoms()
```

```python
f_pi_AD = sp.solve(equations[1].subs({i:i_star}), pi)[0].simplify()
f_AS_curve = sp.lambdify([Y, f, g, E_pi, eps_pi], f_pi_AS)
f_AD_curve = sp.lambdify([Y, i_bar, pi_bar, i_star, alpha, beta], f_pi_AD)
f_sigma_curve = sp.lambdify([Y, G, Y_star, i_star, E_pi, b0, b1, b2, b3, b4], sigma_curve)

f_pi_eq = sp.lambdify([pi_bar, i_bar, i_star, E_pi, eps_pi, alpha, beta, f, g], solutions[pi])
f_Y_eq = sp.lambdify([pi_bar, i_bar, i_star, E_pi, eps_pi, alpha, beta, f, g], solutions[Y])
f_sigma_eq = sp.lambdify([G, E_pi, Y_star, i_star, pi_bar, i_bar, eps_pi, alpha, beta, b0, b1, b2, b3, b4, f, g], solutions[sigma])
```

```python
b0 = 1 # demand constant
b1 = 1.1 # government spending multiplier
b2 = -0.3 # sensitivity of investment to real interest rate
b3 = 0.5 # sensitivty of GDP gap to foreign growth (foreign wealth effect)
b4 = -0.2 # sensitivity of GDP gap to exchange rate (trade balance sensitivity to price changes)

G = 0 # government spending
alpha = 1.5 # central bank sensitivity to excess inflation
beta = 0.8 # central bank sensitivity to excess growth
pi_bar = 2 # central bank inflation target
i_bar = 1 # central bank "base" interest
E_pi = 1 # Expected future inflation

Y_star = 1 # Foreign GDP
i_star = 1 # Foreign interest rate

sigma_bar = 1 # Fixed exchange rate

f = 1 # Phillip's curve parameter
g = 2 # Okun's law parameter
pi_shock = 0 # Supply shock to inflation 
```
